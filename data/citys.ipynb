```python
# read from csv yelp_academic_dataset_restaurants.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import re
import nltk
from nltk.corpus import stopwords

# read from csv
df = pd.read_csv('yelp_academic_dataset_restaurants.csv')
```

```python
# get citys and states
# in the format like {[‘Abington', 'PA’],
#  [‘Abington Township', 'PA’],
#  [‘Affton', 'MO’],
#  etc.

# it's a set of lists

citys = df[['city', 'state']].drop_duplicates().values.tolist()
```

```python
# eliminate the duplicates
citys = list(set([tuple(t) for t in citys]))
```

```python
# give me the citys as a set of lists
citys = [list(t) for t in citys]
```

```python
# give me the unique states
states = list(set([t[1] for t in citys]))
states
```

```python
# tell me what restaurants are in the state of XMS
# XMS is a state

for restaurant in df[df['state'] == 'XMS']['name']:
    print(restaurant)
```

```python
# print out all restaurants' postcode if it's not only consist of numbers

for restaurant in df['postal_code']:
    # if it's float
    if isinstance(restaurant, float):
        continue
    if not restaurant.isdigit():
        print(restaurant)
    # print(restaurant)
```
