```python
# read reviews.csv as csv file

file = open('yelp_academic_dataset_restaurants.csv', encoding='utf-8')
```

```python
import pandas as pd
# read it as a pandas dataframe
restaurants = pd.read_csv(file)
```

```python
# give the list of all the city, state as a tuple (city, state)
def get_city_state_list(restaurants):
    city_state_list = []
    for index, row in restaurants.iterrows():
        city_state_list.append((row['city'], row['state']))
    return city_state_list

len(set(get_city_state_list(restaurants)))
```

```python
set(get_city_state_list(restaurants))
```

```python
# get the set of all the names
def get_name_set(restaurants):
    name_set = set()
    for index, row in restaurants.iterrows():
        name_set.add(row['name'])
    return name_set

len(get_name_set(restaurants))
```
